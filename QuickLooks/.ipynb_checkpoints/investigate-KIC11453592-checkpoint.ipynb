```python
from lightkurve import KeplerTargetPixelFile as ktpf
from lightkurve import KeplerLightCurveFile
import lightkurve
from tqdm import tqdm
```

```python
target = 'KIC11453592'
lc = lightkurve.search_lightcurvefile(target, quarter=6).download().PDCSAP_FLUX.normalize()
for q in tqdm(range(6, 17)):
    try:
        lc = lc.append(lightkurve.search_lightcurvefile(target, quarter=q).download().PDCSAP_FLUX.normalize())
    except:
        pass
lc = lc.remove_nans()
```

```python
lc.plot()
```

There seems to be some periodic variation--- is this physical? Im not sure, so I'm going to take it out and see if it improves things.

```python
flc = lc.flatten()
flc.plot()
```

```python
from astropy import units as u
pg = flc.to_periodogram(freq_unit=u.microhertz)
pg.plot()
```

```python
ax = pg.plot(scale='log')
pg.smooth().plot(scale='log', ax=ax)
```

```python
snr, bkg = pg.flatten(return_trend=True)
ax = pg.plot(scale='log')
bkg.plot(scale='log', ax=ax)
```

```python
snr.plot()
```

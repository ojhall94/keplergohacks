## Lightkurve has a ton of new features, I want to try them out!!

```python
import lightkurve as lk
import matplotlib
matplotlib.rcParams['text.usetex'] = False
```

```python
data = lk.search_targetpixelfile("Kepler-10",cadence='short').download_all()
lc = data[0].to_lightcurve()
for d in data[1:]:
    lc = lc.append(d.to_lightcurve())
```

```python
lc.plot()
flc = lc.flatten().remove_outliers().remove_nans()
flc.plot()
```

```python
pg = flc.to_periodogram()
pg.plot()
```

```python
snr = pg.flatten()
snr.plot()
```

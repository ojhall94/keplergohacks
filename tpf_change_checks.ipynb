```python
import lightkurve as lk
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('paper')
```

```python
searchresult = lk.search_targetpixelfile('Kepler-10', mission='Kepler')
```

```python
tpf = searchresult.download()
```

```python
tpf.plot(column='not a column')
```

```python
tpf.plot(column='FLUX',cadenceno=1341093481039481)
```

```python
fig, axess = plt.subplots(3,2, figsize=(10,10))
axes = axess.flatten()
tpf.plot(column='FLUX',ax=axes[0])
axes[0].set_title('Flux')

tpf.plot(column='FLUX_ERR',ax=axes[1]) 
axes[1].set_title('Flux Err')

tpf.plot(column='FLUX_BKG',ax=axes[2]) 
axes[2].set_title('Flux Bkg')

tpf.plot(column='FLUX_BKG_ERR',ax=axes[3]) 
axes[3].set_title('Flux Bkg Err')

tpf.plot(column='RAW_CNTS',ax=axes[5]) 
axes[5].set_title('Raw Counts')

fig.tight_layout()

tpf.plot(column='COSMIC_RAYS',ax=axes[4]) 
axes[4].set_title('Cosmic Rays')

```

```python
import lightkurve as lk
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('paper')
```

```python
searchresult = lk.search_targetpixelfile('Kepler-10', mission='TESS')
```

```python
tpf = searchresult.download()
```

```python
fig, axess = plt.subplots(2,2, figsize=(10,10))
axes = axess.flatten()
tpf.plot(ax=axes[0])  #Should be image
axes[0].set_title('Flux')
tpf.plot(bkg=True, ax= axes[1]) #Should be background
axes[1].set_title('Background')
tpf.plot(add_bkg=True, ax= axes[2]) #Should be background + image
axes[2].set_title('Flux + Background')
tpf.plot(bkg=True, add_bkg=True, ax= axes[3]) #Should be background + image
axes[3].set_title('Flux + Background')
```
